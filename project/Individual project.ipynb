```python
import csv
f = open ('melanoma.csv', 'rb')
reader = csv.reader (f)
rownum = 0
year = []
incidence = []
index = 0
for row in reader:
    if index != 0:
        year.append(row[0])
        incidence.append(row[1])
    
    index += 1
print year
print incidence
f.close()
```

```python
from pylab import *
```

```python
plot(year, incidence)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title ('Melanoma incidences every five years')
ax.set_xlabel('years')
ax.set_ylabel('Number Of Incidences')
ax.plot(year, incidence)
fig.savefig("number_of_incidences_every_five_years")
```

This plot graph shows the number of incidences every 5 years
